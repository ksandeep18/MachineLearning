```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Basic Plotting
# Create a simple line plot
plt.plot([1, 2, 3, 4], [5, 6, 7, 8])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Simple Line Plot")
plt.show()

# Scatter plot
plt.scatter([1, 2, 3, 4], [5, 6, 7, 8])
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot")
plt.show()


# Bar chart
plt.bar([1, 2, 3, 4], [5, 6, 7, 8])
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Chart")
plt.show()


# Histogram
data = np.random.randn(100)
plt.hist(data, bins=20)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram")
plt.show()


# Box plot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data, labels=['0', '1', '2'])
plt.ylabel("Value")
plt.title("Box Plot")
plt.show()



# Pie chart
labels = 'A', 'B', 'C', 'D'
sizes = [15, 30, 45, 10]
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()


# Customize plots:
# Colors
plt.plot([1, 2, 3, 4], [5, 6, 7, 8], color='red', marker='o', linestyle='dashed', linewidth=2, markersize=8)


#Labels and Legends
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], label='Squared') #add label
plt.plot([1, 2, 3, 4], [1, 2, 3, 4], label='Linear') #add label
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot with Legend")
plt.legend()
plt.show()



# Figure size
plt.figure(figsize=(10, 6))  # Width, height in inches

# Subplots
plt.subplot(2, 2, 1) #2x2 grid, first subplot
plt.plot([1, 2, 3])

plt.subplot(2, 2, 2) #2x2 grid, second subplot
plt.plot([4, 5, 6])

plt.subplot(2, 2, 3) #2x2 grid, third subplot
plt.plot([7, 8, 9])

plt.subplot(2, 2, 4) #2x2 grid, fourth subplot
plt.plot([10, 11, 12])

plt.tight_layout() # Adjust subplot params for a tight layout
plt.show()

#Saving Figures
plt.plot([1, 2, 3, 4], [5, 6, 7, 8])
plt.savefig('my_plot.png') #save the plot


#Working with Pandas DataFrames
data = {'col1': [1, 2, 3], 'col2': [4, 5, 6]}
df = pd.DataFrame(data)

df.plot(x='col1', y='col2', kind='line')
plt.show()

df.plot.bar(x='col1', y='col2')
plt.show()


#Seaborn Integration
import seaborn as sns
# Load an example dataset
tips = sns.load_dataset("tips")

# Create a scatterplot
sns.scatterplot(x="total_bill", y="tip", hue="smoker", data=tips)
plt.show()

sns.countplot(x="day", data=tips)
plt.show()

#More advanced plots and customization options are available in the documentation.

```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load an example dataset
tips = sns.load_dataset("tips")

# 1. Distribution Plots

# Histograms: Visualize the distribution of a single variable
sns.histplot(tips['total_bill'], kde=True) # kde adds kernel density estimate
plt.show()

# Kernel Density Estimate (KDE): Smooth representation of the distribution
sns.kdeplot(tips['total_bill'])
plt.show()

# Box Plots: Show distribution statistics (median, quartiles, outliers)
sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()

# Violin Plots: Combine box plot with kernel density estimate
sns.violinplot(x='day', y='total_bill', data=tips)
plt.show()


# 2. Categorical Plots

# Count Plots: Show counts of observations in each categorical bin
sns.countplot(x='day', data=tips)
plt.show()

# Bar Plots: Show the average value of a numerical variable for each category
sns.barplot(x='day', y='total_bill', data=tips)
plt.show()


# 3. Relationship Plots

# Scatter Plots: Visualize the relationship between two numerical variables
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='smoker', size='size')
plt.show()


# Pair Plots: Show pairwise relationships between multiple variables
sns.pairplot(tips, hue='sex')
plt.show()

# Joint Plots: Show bivariate relationship between two variables along with their marginal distributions
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde') # kind can be 'scatter', 'kde', 'hist', 'reg', 'resid'
plt.show()


# Heatmaps: Visualize correlation matrix
corr = tips.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm') # annot=True shows correlation values, cmap controls color scheme
plt.show()


# 4. Regression Plots

# Reg Plots: Show linear regression between two variables
sns.regplot(x='total_bill', y='tip', data=tips)
plt.show()

# lmplot: More flexible version of regplot, supports faceting
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', hue='smoker')
plt.show()


# 5. Customization

# Color palettes: Use different color schemes
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='smoker', palette='viridis')
plt.show()

# Styles and contexts: Adjust visual elements for presentation
sns.set_style("whitegrid") # Set style
sns.set_context("paper")  # Set context (paper, notebook, talk, poster)
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.show()

# Figure size and aspect ratio
plt.figure(figsize=(12,6))
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.show()

# Axis labels and titles
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.xlabel('Total Bill Amount')
plt.ylabel('Tip Amount')
plt.title('Relationship Between Total Bill and Tip')
plt.show()
```
